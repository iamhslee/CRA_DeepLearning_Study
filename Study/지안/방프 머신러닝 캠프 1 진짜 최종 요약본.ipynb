# 한동대 머신러닝 캠프 
## Introduction to Machine Learning 

AI, Machine Learning, Deep Learning의 관계 
- 인공지능 > 머신러닝 > 딥러닝 순의 집합관계를 이룸
- 머신러닝은 인공지능의 분야(approach) 중 하나이며, 딥러닝은 머신러닝의 분야(approach) 중 하나이다.

AI
- 컴퓨터 소프트웨어를 이용해서 지능을 구현하는 것(strong AI와 관련)
- 지능이 필요한 일을 자동화하기 위한 기술(Week AI와 관련)
- 인공지능의 정의
    1) strong AI, AGI
        + 어떤 입력이 들어왔을 때 지능이 지능적인 결과를 낸다.
        + 한마디로 인간 수준의 AI이다. 
        + 인간처럼 학습을 받으며, 모든 분야를 인간처럼 배우게 된다.
    2) Week AI, Narrow AI
        + 실제론 지능이 없지만, 지능이 요구하는 결과물을 내면 된다.
        + 특정 임무만을 수행하는 인공지능을 의미한다.(예: 알파고 등) 

AI와 일반적인 프로그램의 차이 
- 일반적인 소프트웨어는 미리 정해진 알고리즘을 따라서 특정 업무를 수행한다.
- 하지만 AI는 
    1) 절차를 정해서 수행하기 어려운 일을 함. 
    2) 환경이 바뀌어도 스스로 적응함.
    3) 불확실성이나 모호함이 존재하는 상황에서 결정을 내림.(예: 바둑을 두는 것)

Machine Learning 
- 어떤 업무를 하는데 있어서 수행 방법을 가르쳐주지 않아도 데이터로부터 그 방법을 스스로 배운다. 
- 인공지능 agent한테 다양한 입력 데이터(사진, 목소리, 병의 증상 등)를 줄 수 있다. 그 입력에 대해서 적절한 결과값을 낼 수 있다면, 그게 지능을 필요로 하는 일을 수행하는 모듈을 만들어낼 수 있다. 
- agent가 하는 일은 입력이 들어왔을 때 어떤 출력을 내느냐로 정의할 수 있다.
- 머신러닝을 할 때는 입출력의 과정이 특정한 수학적 틀을 기반으로 한다.

- Connection weights이 parameters/structure에 해당하는 부분이며, 이게 Training에 의해 최적화된다. 

머신러닝의 종류
1. Supervised learning(지도학습)
    - 정답이 있는 데이터를 활용해 데이터를 학습시키는 것
    - 입력값이 주어지면, 입력값에 대한 Label(y data)을 주어 학습시킨다.
    - 예: 고양이 사진을 입력값으로 주고 해당 사진에 대한 label 고양이란 정보를 제공한다.
    
    
2. Unsupervised learning(비지도학습)
    - 정답 라벨이 없는 데이터를 비슷한 특징끼리 묶어 새로운 데이터에 대한 결과를 예측한다.
    - 비지도 학습이 할 수 있는 것
        1) clustering: 비슷한 데이터끼리 묶는다. 특징 분석이나 데이터의 패턴이나 분포가 무엇인지를 알아내는데 중요한 역할을 한다.
        2) reproduction: 어떤 데이터를 제공받으면, 그 데이터에 대한 중요한 특징을 기억한 뒤 다시 그 데이터와 비슷하게 출력한다.
        3) latent variable models(hidden factor analysis): 어떤 현상을 관찰한 뒤 그 현상이 발생한 숨겨진 이유를 역으로 추적함. 
         
        - 이 예시처럼 흰색 증상을 바탕으로 검은색 원인을 파악한다.
        
3. Semi-supervised learning(반지도학습)
    - 수집한 데이터에 정답을 부여하는 것 즉 labeling된 데이터의 부재를 극복하기 위해 나왔다.
    - 
    - 위 사진처럼 레이블된 데이터 주변에 레이블되지 않은 데이터를 추가함으로써 좀 더 구분선을 쉽게 정할 수 있도록 한다.
    - 예를 들어 레이블된 개, 고양이 데이터를 갖고 있지만, 그 수가 적다면 개, 고양이에 대한 구분이 명확히 안 될 수 있다. 이를 해결하기 위해 레이블되지 않은 데이터라도 구해서 데이터의 수를 늘려야 한다. 레이블되지 않은 데이터라 함은 쉽게 말하면 인터넷에 개, 고양이를 검색해서 나온 사진이라 생각하면 된다. 실제로 개, 고양이에 대한 사진이 많이 나오지만, 가끔 엉뚱한 사진이 나올 때도 있다. 이렇게 모은 데이터를 기존 데이터에 추가해서 분포시키면 좀 몇몇 오류가 있을 수 있어도 데이터가 훨씬 많아졌기 때문에 비교적 정확하게 개와 고양이를 구분할 수 있다. 
    
4. Reinforcement learning(강화학습)
    - 인공지능 agentr가 주변 환경하고 상호작용을 한다고 가정한다. 그래서 이 agent가 어떻게 주변 환경과 가장 잘 상호작용하게 할 지를 구한다.
    
    - 위의 사진을 설명하고자 예를 들면, 바둑을 두는 알파고를 agent라고 한다. 그리고 environment은 바둑판이다. agent는 각 상태(현재 바둑판의 국면)에서 action(행동)할 수 있다. 이 environment는 agent에 reward(집을 지은 것)를 할 수 있다. 주로 승패를 본다.따라서 agent가 해야 하는 것은 rewards를 극대화하는 것이다.
    - 강화 학습에 사용되는 용어
        1) Policy: 각 state가 주어졌을 때 해야 되는 행동에 대한 확률(조건부 확률로 표현됨)
        
        2) Value of state: 바둑판을 예로 들면, 현 바둑판의 국면이 끝까지 갔을 때 이길 수 있는 확률
        3) Value of action (Q-function): 특정 state에서 어떤 행동을 했을 때 그 다음에 이길 확률
   - 어떤 행동이 좋은 결과를 야기하면 그 행동의 확률을 높이고, 반대일 경우 그 확률을 낮춘다. 

Alpago 
- 기존에 바둑을 둘 때는 Tree Search 기법을 통해 바둑판에 둘 수 있는 경우의 수 중 최적의 수를 찾아 돌을 두었다. 하지만 이 기법은 고려해야 할 경우의 수가 너무 많다는 문제점을 갖고 있다. 
- 따라서 효율적으로 알고리즘을 사용하기 위해서는 탐색할 경우의 수를 좁힌 다음 그 범위만 탐색한다. 
- 위의 행동은 2가지 network를 통해 작동된다.
    1) Policy network 
        + 입력 데이터는 바둑판의 현재 국면이 되며, 다음 수를 어디 두는 게 좋은지 확률을 계산한 뒤 그 확률을 바둑판에 막대 그래프 형식으로 표시한다. 그 표시된 그림을 바탕으로 확률이 비교적 높은 수들을 다음 수로 활용한다.  
        + 계산해야 할 경우의 수를 줄여준다는 장점이 있다.
- 
    2) Value network 
        + 해당 국면에서 바둑을 계속 둘 경우 이길 확률이 얼마일지 계산한다.
        + 계산해야 할 단계를 줄여준다는 장점이 있다.
- 
- 즉 Policy network를 통해 착수할 다음 돌의 경우의 수를 줄여주고, Value network를 통해 그 돌을 두었을 때 이길 확률을 계산한다.

Recognition System
- 인공지능으로 푸는 대표적인 문제 2가지
    1) Classification(분류)
        + 어떤 입력 데이터가 들어오면 그 입력의 범주(category)를 예측한다.
        + 예: 어떤 인물이 남자인지 여자인지 맞추는 것
    2) Regression(회귀)
        + 파산 예측이나 판매 예측처럼 어떤 일이 일어날 확률을 예측한다. 
        + 종속 변수(Independent variables)가 주어지고 그것을 통해 독립 변수(target/Dependent variables)를 예측한다. 
        + 예: 키와 성별을 통해 체중을 맞추는 것 
        + 회귀는 아래와 같은 함수로 표현할 수 있다.
    -  
        + y => 출력 변수이자 입력 변수에 대한 함수, 세타 => 매개 변수(ex. ax + b에서 a와 b 같은 존재 => 훈련 데이터에 의해 학습된 매개변수는 훈련 데이터를 정답에 맞게 혹은 가깝게 구별 혹은 예측할 수 있다.), X => 입력 데이터
        + regression 함수의 종류
    -  
        + logistic 회귀는 참, 거짓(1 또는 0)을 판별한다.
- 
- 위 사진과 같은 방법으로 recognition system이 작동한다. 
    1) Target: 실제 세계에 존재(사람, 나무, 동물 혹은 어떤 현상 등)
    2) Sensing: 위 타겟을 카메라를 찍거나 마이크로 녹음하는 등의 행위를 통해 아날로그의 데이터를 디지털의 데이터로 변환한다. 
    3) Preprocessing/Detection/Segmentation: 변환한 데이터 중 사용할 데이터만 있도록 처리한다.
    4) Feature Extraction: 인공지능 학습을 위해 입력 패턴(데이터)을 숫자 벡터로 바꾼다.
    5) Classification(혹은 Regression): 위의 벡터를 인공지능에 집어 넣어서 학습도 시키고 분류 혹은 회귀 작업을 수행한다. 
        + Model의 매개 변수들은 훈련 데이터를 통해 조정된다. 
- 왜 입력 패턴을 벡터로 바꾸어 사용하는가?
    1) 데이터를 다루기가 쉬워진다. 
        + 거의 대부분의 머신러닝 알고리즘(수학적인 방법론)이 입력 데이터가 벡터라고 가정하는 경우가 많다. 
        + 영상의 크기나 사진의 크기는 모두 다를 수가 있는데, 이 데이터들이 고정된 크기의 벡터로 들어와야 
    2) 중요한 정보만 추출해서 데이터의 크기를 줄일 수 있다.
        + 사람 얼굴을 인식하는 데, 눈, 코, 입의 개수보다는 위치가 중요함으로 중요한 것만 숫자로 바꾸기 위해서 벡터를 사용한다. 
    3) 효율성
        + 중요한 데이터만 숫자로 뽑아서 활용하면 시간을 단축할 수 있다.  

Complexity and Generalization Ability 
- 
- 오른쪽 그림과 같이 그 오분류된 데이터까지 다 제대로 분류하기 위해 경계선을 복잡하게 만들면 더 큰 오류를 발생시킬 수 있다. 
- 위의 데이터 분포는 훈련 데이터로만 이뤄져 있는데, 소수의 오분류된 훈련 데이터를 신경쓰다간 데이터의 분류 기준이 망가져 새로 들어온 데이터를 제대로 분류하지 못할 수 있기 때문이다.  
- 이런 경우를 overfiting(과적합)이라고 부르며, 모델이 굉장히 복잡할 때(데이터가 적은 상태에서 layer가 많은 모델을 사용하는 것 등) 발생한다.  
- overfiting은 training 데이터의 오류율(Error rate)가 감소하고 테스트 데이터의 오류율이 증가하는 현상을 통해 발견할 수 있다.

Bayesian Theorem
- 
- 위의 그림을 '키와 몸무게'와 '성별'의 관계를 통해 설명하겠다.
- 키와 몸무게(X)로 성별(w)을 구분한다고 했을 때 X를 기준으로 w을 찾는 것은 힘들다. 일일이 X에 해당하는 사람을 모아 w의 비율을 따져야 하기 때문이다.즉 위의 수식을 구하기 힘들다는 뜻이다.
- 반면에 w를 일정 수준 모은 뒤 그 안에서 X의 분포를 찾는 것은 쉽다. 그냥 나열하면 되기 때문이다.
- 즉 이 베이지안 이론은 아래 그림처럼 측정하기 어려운 확률을 쉬운 확률로 바꿔 계산하는 것이다.
